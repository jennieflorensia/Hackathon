```python
import pandas as pd
import numpy as np
from matplotlib import cm

rm = pd.read_csv('epi_r.csv')
```

```python
rm[['calories', 'protein', 'fat', 'sodium']] = rm[['calories', 'protein', 'fat', 'sodium']].fillna(rm[['calories', 'protein', 'fat', 'sodium']].median())

rm.info()
```

```python
rm.isna()
```

```python
from sklearn.model_selection import train_test_split
X = rm.drop(columns = ['title','rating','calories', 'fat', 'protein', 'sodium','#cakeweek','#wasteless', '22-minute meals', '3-ingredient recipes', '30 days of groceries', 'advance prep required'])
y = rm['title']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=45)

print ("Train size : ",X_train.shape)
print ("Test size : ",X_test.shape)
```

```python
print(X)
```

```python
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()

gnb.fit(X_train, y_train)
```

```python
hasil = gnb.predict(X_test)
print(hasil)
```
